```python
# Loading the data to be merged

credits = pd.read_csv('credits.csv')
```

**Preparing the data for merging**

```python
# Looking if we have 'id' duplicates

credits.id.value_counts()
```

```python
credits[credits.duplicated(subset=['id'], keep=False)].sort_values('id')
```

```python
credits.drop_duplicates(subset='id', inplace=True)
```

```python
credits.id.value_counts()
```

**Merging the data (Left Join)**

```python
# Checking if there are movies who is not (~) in the credits dataset

df[~df.id.isin(credits.id)]
```

```python
# Looking if there are 'id's in the credits dataset that aren't in the movies dataset

credits[~credits.id.isin(df.id)]
```

```python
# Merging both datasets

df = df.merge(credits, how='left', left_on='id', right_on='id')
```
