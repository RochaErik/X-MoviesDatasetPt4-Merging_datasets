**Preparing the data for merging**

```python
# Looking if we have 'id' duplicates

credits.id.value_counts()
```

```python
credits[credits.duplicated(subset=['id'], keep=False)].sort_values('id')
```

```python
credits.drop_duplicates(subset='id', inplace=True)
```

```python
credits.id.value_counts()
```

**Merging the data (Left Join)**

```python
# Checking if there are movies who is not (~) in the credits dataset

df[~df.id.isin(credits.id)]
```

```python
# Looking if there are 'id's in the credits dataset that aren't in the movies dataset

credits[~credits.id.isin(df.id)]
```

```python
# Merging both datasets

df = df.merge(credits, how='left', left_on='id', right_on='id')
```

```python
df
```

```python
df.info()
```

**Cleaning and transforming the 'cast' column**

```python
import ast
import numpy as np
```

```python
df.cast[0]
```

```python
df.cast = df.cast.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else np.nan)
```

```python
pd.DataFrame(df.cast[0])
```

```python
# Creating a new column with the cast size

df['cast_size'] = df.cast.apply(lambda x: len(x))
```

```python
df.cast_size
```

```python
# Analizing the most comum cast size

df.cast_size.value_counts(dropna=False).head(25)
```

```python
# Extracting the actors name for each movie and storing separated by a '|'

df.cast = df.cast.apply(lambda x: '|'.join(i['name'] for i in x) if isinstance(x,list) else np.nan)
```
